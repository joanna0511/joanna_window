```python
import requests
from  requests import JSONDecodeError
from pprint import pprint

def download_json()->dict[any]:
    bike_url = 'https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json'
    try:
        response = requests.get(bike_url)
    except Exception:
        raise Exception("連線失敗")
    else:
        if response.status_code == 200:
            try:
                all_data:list[any] = response.json()
                return all_data
            except JSONDecodeError:
                raise Exception("api_key為測試用,連線已至上限,請稍後再試")
        else:
            raise Exception("下載狀態碼不是200")
```

```python
try:
    all_data:list[any] = download_json()
except Exception as error:
    print(error)
```

```python
from pydantic import BaseModel, RootModel, Field

class Info(BaseModel):
    sna:str
    sarea:str
    mday:str
    act:str
    updateTime:str
    total:int
    available_rent_bikes:int
    latitude:float
    lon:float = Field(alias='longitude')
    available_return_bikes:int


class Records(RootModel):
    root:list[Info]
ubike_records:Records = Records.model_validate(all_data)
data = ubike_records.model_dump()
```
